```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import pandas as pd
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import Select
import openpyxl
import PyPDF2, os
from pypdf import PdfMerger
from PyPDF2 import PdfFileMerger
import time



chrome_options = Options()
download_dir = os.getcwd()
chrome_options.add_experimental_option('prefs', {
    'download.default_directory': download_dir,
    'download.prompt_for_download': False,
    'download.directory_upgrade': True,
    'safebrowsing.enabled': True
})

x_path = input("Enter the path of Excel File : ")


for filename in os.listdir(os.getcwd()):
    if filename.startswith("individual-file") and filename.endswith(".pdf"):
        os.remove(filename)
        
        
data=pd.read_excel(x_path,header=None)
data=data.dropna()

browser=webdriver.Chrome(options=chrome_options)
browser.get("https://sbss.dewii.in/sbss/admin/login")
browser.implicitly_wait(5)
browser.find_element(By.CLASS_NAME,"form-control").send_keys("admin")
browser.find_element(By.XPATH,"/html/body/div/div[2]/form/div[3]/div/input").send_keys("Sbss@#12321")
browser.find_element(By.XPATH,"/html/body/div/div[2]/form/button").click()
browser.implicitly_wait(5)

browser.find_element(By.XPATH,'//*[@id="navmenu"]/li[11]/a').click()
browser.implicitly_wait(5)

print("Length of data : ",len(data))
for i in range(1,(len(data))):
    try:
        workmen_pf_no=data[1][i]
        workmen_month=str(data[2][i])
    
    
        browser.find_element(By.XPATH,'//*[@id="individual"]/form/div/div[1]/div[1]/span/span[1]/span/span[2]').click()
        browser.find_element(By.XPATH,'//html/body/span/span/span[1]/input').send_keys(workmen_pf_no)
        browser.find_element(By.XPATH,'//html/body/span/span/span[1]/input').send_keys(Keys.ENTER)



        browser.implicitly_wait(5)
        browser.find_element(By.XPATH,'//*[@id="individual"]/form/div/div[2]/div[1]/input').click()
        browser.find_element(By.XPATH,'//*[@id="individual"]/form/div/div[2]/div[1]/input').send_keys(workmen_month)
        browser.find_element(By.XPATH,'/html/body/div[3]/div[2]/table/tbody/tr/td/span[4]').click()
        browser.implicitly_wait(5)

        browser.find_element(By.XPATH,'//*[@id="individual"]/form/div/div[3]/div/button').click()
        browser.find_element(By.XPATH,'//*[@id="cmd"]').click()
        time.sleep(3)
    except:
        pass
    
browser.close()



x = [a for a in os.listdir() if a.endswith(".pdf")]
 
merger = PdfFileMerger()
 
for pdf in x:
    merger.append(open(pdf, 'rb'))
 
    with open("result.pdf", "wb") as fout:
        merger.write(fout)












# pdfiles = []

# for filename in os.listdir('.'):
#         if filename.endswith('.pdf'):
#                 if filename != 'merged.pdf':
#                         pdfiles.append(filename)
                        
# pdfiles.sort(key = str.lower)


if os.path.exists("result.pdf"):
    os.remove("result.pdf")
    merger.write("result.pdf")
    merger.close()

time.sleep(2)
for filename in os.listdir():
    if filename.endswith('.pdf'):
        os.remove(filename)

        


  








```
